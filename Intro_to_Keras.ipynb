This notebook is going to go through most of what a typical data science project pipeline looks like.

**1. Exploratory Data Analysis**
  
  This is the first step of any machine learning task, actually looking at the data we have. This can (and should) be done visually, as in opening the images and taking a look at them; but also:
  - Looking at the label distribution (is the dataset balanced or not? (are there as many 1s as 0s))
  - Can we see any difference between images with label 1 and label 0? (this can be visually, but also spectrally (deeper look into each individual bands))
  - Can we find any outliers, and if so, what are they charaterized by?
  - How similar / different are the images from one another?
 
Basically the goal here is to get an **understanding** of the data we have. To go from the images and labels, to the actual problem we are trying to solve. Domain knowledge usually helps, this is where additional research on the topic is done (here the domain knowledge is oil palm trees, what they look like, where they are planted, etc. as opposed to doing research on the machine learning itself).
 
 This is particularly true for competitions, where we directly dive into data. In a 'real-life' scenario, the data science team is most of the time part of the data collection process, and so has already an understanding of the data it applies a model to. However, it is important to keep in mind elements like having a balanced dataset, looking for outliers, etc.
 

**2. Understanding the task**

  The problem we are trying to solve is the decision factor on what kind of model we are going to implement, which _tools_ to use. A classification task is very different different than a segmentation one. Even for classifciation tasks, having 2 classes or 20 is very different. Sometimes one image can also belong to multiple classes, that's even different.
  
  At this stage, we usually look up machine learning models, how to implement them, what are the benefits / drawbacks of each model. The goal is to get an idea of what we are going to do, before we even start writting a line of code. Each problem is different, but some are also similar, and a solution for one problem might be a good start for a similar problem. No need to reinvent the wheel.
  
  It's usually at this moment that a loss function and a metric (or more than one by the way) are chosen, that will allow to evaluate the performance of the model. A classification task should be evaluated differently than an object detection problem, which itself is different than a segmentation task.
  
**3. Pre-processing data**

  This can be sort of optional, yet helps a lot to preprocess our images. It can be done in a lot of different ways, with removing the outliers, resampling images to the same shape / format, doing data augmentation, etc.
  
  This is important as models learn best when data is cleaned.
  
**4. Making the model**
  
  Now, this is where the actual learning happens. Plugging into the matrix and implementing what we think would be a good start. (the acual model itself can have as little as 5 lines or even less, so this really isn't the biggest part of the work by the way)
  
  Data science is (still) a lot of trial an error, with experience and knowledge helping a lot what works and what doesn't. This means at first, we want to make a very basic model, and gradually make it better and more complex (if needed), specifically tailored for the problem we are trying to solve.
  
  One very [good advice I read from by Andrej Karpathy](https://twitter.com/karpathy/status/1013244313327681536?s=20) (director of AI at Tesla, previous researcher at OpenAI) is to overfit as early as possible. Taking a small sample of data and actually making a pipeline that works, with very little regard to the output performance, it should just output _something_ that is not complete noise. Once that is done, then we know we can scale it up, adding more features, and fine-tuning it.
  
  Working on a small sample of dataset allows for very rapid prototyping (training in a few minutes instead of hours / days).
  
**5. Error Analysis**

  Once the model is trained, it's time to make some predicitons with it.
  
  Spoiler alert: it's going to make mistakes.
  
  Even if it get's an accuracy of 100%, then something is most probably wrong. That's just not possible, you are probably overfitting, or have a model that will most likely do if shown new data.
  
  Error Analysis simply means looking into the predictions made by the model to investigate what it got wrong, but also what it got right. There are lots of different way to do this, and they are specific to each task. The insight we gain from looking into this will allow us to improve the model, but also help us better clean the data, find other outliers, etc.
  
**6. Go back to step 1**
(_until we run out of time, money, or that we have solved the problem to an acceptable level_)

Depending on the sources of error, we could need to better prepare the data, use a different model, change the loss function, or even try to get more data if possible.

While we can't take a look at all the images individually, we can have a look at the pixel values of each image and see if they differ from one class to the other.

(This is based on the [this notebook from Pablo Gómez](https://www.kaggle.com/gomezp/complete-beginner-s-guide-eda-keras-lb-0-93) on a previous Kaggle competition)

```
def load_img(nb_images, df):
  '''
  Loading nb_images of images into an array
  '''
  X = np.zeros([nb_images, 256, 256, 3], dtype=np.uint8)
  for i, row in tqdm_notebook(df.iterrows(), total=nb_images):
    if i==nb_images:
      break
    
    X[i] = cv2.imread('train_images/'+row['image_id'])
  
  return X
```

# __2. Understanding the task__

This is a binary classifcation task. The goal is simply to say, for a given image, if there is oil palm in it or not. 

For this, a lot of models exist. We will start with some pre-trained models with the most common architectures that are used for these types of classifications.

Especially when it comes to deep learning, there are a lot of different types of architectures, to do image classifciation, but also speech recognition, text understanding, etc. Each problem has a different set of types of models. For most of the computer vision tasks (that is, allowing a computer to 'see', so everything to do with images) we use mostly Convolutional Neural Networks (CNNs). If you feel like reading, [go ahead](http://cs231n.github.io/convolutional-networks/) there's a lot to learn!

Inside the convolutional neural network family, there are many types of architectures that have been found over the years that do well (or not) at specific tasks. Classifying an image is not the same as trying to localize an object in an image.

We will be using a [pre-trained ResNet50](https://www.kaggle.com/keras/resnet50) as a start to see what we can get with that. It is a pretty good, and standard model used for object classification. [See build-in implemtation in Keras].(https://www.kaggle.com/keras/resnet50)

As for the loss, [binary cross-entropy](https://towardsdatascience.com/understanding-binary-cross-entropy-log-loss-a-visual-explanation-a3ac6025181a) as it is a go-to for binary classification tasks.

The actual metric we are trying to maximize however is the AUC, or [Area Under the Curve](https://www.kaggle.com/learn-forum/53782). This takes into account true and false positives.

# __3. Pre-processing the data__

This step is pretty important, as we are not going to directly feed in the images to our model. We could, but that wouldn't work very well.

Probably the most important thing we could do, is to normalize the images we input in the model. That means switching the pixel value range from [0, 255] to [0, 1]. This is simply done because neural network models are designed around values ranging from 0 to 1.

One of the biggest areas where we can improve the results of our models is with what is called data augmentation, which consists in artifically creating more data by flipping, zooming in, shifting, etc. the images to create 'new' images.

The batch size is the number of images that will be trained at a time. We need to split the data in 3:
- A **training** set: will be used to train the data, the biggest set, usually from 80, to 95-99% of the data depending on what we have, and how much we have
- A **test** set: In our case, this is the images we actually want to make the prediction on. In real life conditions, this would be the data that is fed into the model once it is set into production, the data that you actually _use_ the model on
- a **validation** set: used to evaluate the performances of the model. We retain part of the data that we have access to during the elaboration of the model. Once our model is trained on the training set, we give it this validation set, but without giving it the labels. We then compare the predicitons with the labels to evaluate the model.

```
train_val_ratio = 0.2   # Ratio of images to be kept in the validation set.
train_test_ratio = 0.2  # Ratio of remaining images that we will use as our test set.

# Splitting the images into a training / validation set
train_X, val_X, train_y, val_y = train_test_split(sample_df,
                                                  sample_df['has_oilpalm'],
                                                  test_size = train_val_ratio,
                                                  random_state = seed_value)

# Taking the remaining training set, and taking a test set from that
test_set_size = np.int(train_X.shape[0]*train_test_ratio)
print('test set size: ', test_set_size)

train_X = train_X[:-test_set_size]
train_y = train_y[:-test_set_size]
test_set = train_X.tail(test_set_size)
```

```
print('Train X shape: ', train_X.shape)
print('Valid X shape: ', val_X.shape)

print('Train y shape: ', train_y.shape)
print('Valid y shape: ', val_y.shape)

print('Test set shape: ', test_set.shape)
```

We now need to create data generators that will feed the data into the model. Since we can feed everything in the VRAM at once, we need to do it by chunks, or batches. However, we don't want to feed the same images twice, or not forget to feed in some, thus requiring the need of a data generator.

We need one for the training, one for validation and finally we will make one for the test set. They are all different because the images are different. We don't want the model to see images that come from the test set during testing, that would beat the purpose of the test set in the first place.

```
train_generator=datagen.flow_from_dataframe(dataframe=train_X,
                                            directory="train_images/",
                                            x_col="image_id",
                                            y_col="has_oilpalm",
                                            subset="training",
                                            batch_size=BATCH_SIZE,
                                            seed=seed_value,
                                            shuffle=True,
                                            class_mode="binary",
                                            target_size=(256, 256))

valid_generator=datagen.flow_from_dataframe(dataframe=val_X,
                                            directory="train_images/",
                                            x_col="image_id",
                                            y_col="has_oilpalm",
                                            subset="validation",
                                            batch_size=BATCH_SIZE,
                                            seed=seed_value,
                                            shuffle=True,
                                            class_mode="binary",
                                            target_size=(256, 256))
```

```
test_datagen=ImageDataGenerator(rescale=1./255.)

test_generator=test_datagen.flow_from_dataframe(dataframe=test_set,
                                                directory="train_images/",
                                                x_col="image_id",
                                                y_col=None,
                                                batch_size=BATCH_SIZE,
                                                seed=seed_value,
                                                shuffle=False,
                                                class_mode=None,
                                                target_size=(256, 256))
```

# __4. Making the model__

As talked about previously, we are going to take a pre-existing model architechure, which has been pre-trained with existing weights, and use that as the core of our model. We are simply going to allow training of the last layers, but freezing all the weights of all the previous layers.

This is known as transfer learning, and is a very common way of starting, which has been shown to yield very good results, especially with limited data

```
input_tensor = Input((256, 256, 3))
# Here is where we give the pre-trained model
base_model = ResNet50(include_top = False, pooling='avg', input_tensor = input_tensor)
x = base_model(input_tensor)
# We simply add a basic dense layer, that ouputs a single number, between 0 and 1, the predicted probability that the image has oil palm
out = Dense(1, activation='sigmoid')(x)

model = Model(input_tensor, out)
model.summary()
```

There 23.5 million trainable parameters. That's why we want a GPU when doing deep learning. That's also why this is done on Colab, we can have access to K80 for free!

Now, we are going to add some [callbacks](https://keras.io/callbacks/). Those are basically features that allow to make the most out of the training.

The ones we are going to use are:
- Model Checkpoint: this will keep the best weights that the model has seen during each epoch. This means, that if the performances gets worse after some time, we will keep the model where the performance was highest
- Early stopping: if we see that the performance keeps going down, after a while we just stop the training. This prevents the training from going on when we have past the optimum point, saving a lot of time.
- Reduce learning Rate on Plateau: If the training does stops imporoving, we reduce the learning rate. The theroy behind it is that we are approaching a converging point, and that we might be taking steps that are too big to reach it.

```
#Callbacks
best_model_weights = './base.model'

checkpoint = ModelCheckpoint(
    best_model_weights,
    monitor = 'val_loss',
    save_best_only = True,
    mode='min',
    verbose=1,
    save_weights_only=False,
    period = 1)

earlyStop = EarlyStopping(
    monitor = 'val_loss',
    min_delta=0.01,
    patience = 5,
    verbose = 1,
    mode = 'min')

reduce = ReduceLROnPlateau(monitor='val_loss',
                           factor=0.5,
                          patience=3,
                          verbose=1,
                          mode='min'
                          )

tbc=TensorBoardColab()

callbacks = [checkpoint,
             earlyStop,
             #TensorBoardColabCallback(tbc), # Comment to gain a few seconds of training
             reduce]
```

If you leave tensorboard uncommented, you get a link, which gives you acces to [Tensorboard](https://www.tensorflow.org/tensorboard/get_started) which is a pretty handy tool to visualize metrics from the model, as well as it's architecture. On the top left, where it says inactive, there's a scroll down menu, go to scalar to see the loss and accuracy graphs in real time.

It's maybe a bit overkill in this context, but it's still a nice tool to know about. It is particularly handy if you are trying to train multiple models to see which one fits your problem better. It's also good to see the models performance while training. It can provide some useful information, to decide to stop a training early if it isn't going well for example.

```
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size

print(f'train step size: ', STEP_SIZE_TRAIN)
print(f'valid step size: ', STEP_SIZE_VALID)
print(f'test step size: ', STEP_SIZE_TEST)
```

```
# It takes about 5 minutes to train the model as it is now

model.compile(optimizer = optimizers.Adam(lr=0.001),
             loss=binary_crossentropy,
             metrics=['accuracy'])

history = model.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=nb_epochs,
                    callbacks=callbacks)
```

One thing that is important to keep in mind is that this is a notebook. So if you rerun the previous cell multiple times, it continues training from where it stopped. So basically you're adding 5 epoch of training each time you rerun it. This is important to keep in mind, as it is one of the traps of using notebooks.

```
# Predicting the class of images in the test set
test_generator.reset()
pred = model.predict_generator(test_generator,
                                     steps = STEP_SIZE_TEST + 1,
                                     verbose=1)
```

```
pred = np.ravel(pred)
print(f'There are {len(pred)} predictions')
pred[:10]
```

These predicitons are from 0 to 1. We need to threshold them to attribute them to either 1 or 0. The simplest way is to say everything >= 0.5 is 1, and everything <0.5 is 0.

However, sometimes there's a better thresholding to make, and it's fairly easy to make a loop that tries different values of thresholds and computes the AUC, and then keeps the best.

```
threshed_pred = {}
for thresh in range(30, 70, 5):
  thresh = thresh / 100
  threshed_pred[thresh] = np.where(pred<thresh, 0, 1)
```

Let's take a look at the Reciever operating characteristic (ROC) and Area Under the Curve (AUC) of the ROC, to see how the model performs for the metric that we actually care about and are trying to maximize: the AUC.

```
from sklearn.metrics import roc_auc_score

AUC_thresh = {}
for thresh in threshed_pred:

  AUC = np.round(roc_auc_score(test_set['has_oilpalm'].ravel(), threshed_pred[thresh]), 4)
  AUC_thresh[thresh] = AUC
  #print(f'AUC at threshold {thresh} is: {AUC}')
  
best_thresh = max(AUC_thresh, key=AUC_thresh.get)
print(f'Highest AUC is: {AUC_thresh[best_thresh]} for thresh {best_thresh}')
```

```
from sklearn.metrics import roc_curve, auc

# ROC
false_positive_rate, true_positive_rate, thresholds = roc_curve(test_set['has_oilpalm'].ravel().astype(int), threshed_pred[best_thresh])

# AUC
auc = auc(false_positive_rate, true_positive_rate)

print(false_positive_rate)
print(true_positive_rate)
print(auc)
```

```
#Plotting ROC and giving the AUC
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(false_positive_rate, true_positive_rate, label='Area Under Curve= {:.4f}'.format(auc))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```
